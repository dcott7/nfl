```python
from sqlalchemy.orm import sessionmaker
from sqlalchemy import create_engine
```

```python
from db.models import Competitor, Event, Team
```

```python
engine = create_engine('sqlite:///sports.db')
Session = sessionmaker(bind=engine)
session = Session()
```

```python
panthers_events = (
    session.query(Event.name)
    .join(Competitor)
    .join(Team)
    .filter(Team.name == "Carolina Panthers")
    .all()
)

# Print event names
for event in panthers_events:
    print(event.name)

# Close session
session.close()
```
